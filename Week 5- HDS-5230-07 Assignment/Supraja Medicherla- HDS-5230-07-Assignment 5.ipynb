```python
import dask.dataframe as dd
import pandas as pd
import numpy as np
```

1. Create a new dataframe object that consists of samples (i.e., rows) corresponding to states in the US.Create a new dataframe object that consists of samples (i.e., rows) corresponding to states in the US.

```python
data = pd.read_csv("timeseries.csv", dtype={'name':str,'level':str,'city':object,'county':str,'state':str,'country':str,'population':np.float64,'lat':float,'long':float,'url':str,'aggregate':object,'tz':str,'cases':np.float64,'deaths':np.float64,'recovered':np.float64,'active':np.float64,'tested':np.float64,'hospitalized':np.float64,'hospitalized_current':np.float64,'discharged':np.float64,'icu':np.float64,'icu_current':np.float64,'growthFactor':np.float64,'date':object})
df = dd.from_pandas(data)

filtered_df = df.query("country == 'United States'")
filtered_df = filtered_df.groupby("state").count()
filtered_df.compute()
```

```python
new_column_df = df.query("country == 'United States'")\
    .assign(seen=lambda x: x.recovered + x.active)
new_column_df.compute()
```

Interpretation: Here, I filtered the dataset to include only records where the country is the United States using the query function. Since the dataset contains global COVID-19 data, this step ensures that only U.S.-specific data is considered for analysis. Next, it creates a new column called seen, which is the sum of recovered and active cases. This metric helps in understanding the total number of observed cases, excluding deaths, and can be useful in analyzing trends in recovery versus active cases. Since Dask operates in a lazy evaluation mode, the .compute() function is used to execute the computations and return a tangible result as a Pandas DataFrame. 

2. During the time period 2020-Jan-01 to 2021-Feb-28, rank states in terms of their per-capita mortality? Compute per-capita mortality during a specific period as the ratio of total deaths during that time period, to the average population of the state (compute the average population during the time period).

```python
# Convert date column to datetime
new_column_df["date"] = dd.to_datetime(new_column_df["date"])

# Filter for the specified time period
time_period_df = new_column_df[
    (new_column_df["date"] >= "2020-01-01") & (new_column_df["date"] <= "2021-02-28")
]

# Group by state and compute total deaths and average population
state_stats = time_period_df.groupby("state").agg(
    total_deaths=("deaths", "sum"),
    avg_population=("population", "mean")
).compute()

# Compute per-capita mortality
state_stats["per_capita_mortality"] = state_stats["total_deaths"] / state_stats["avg_population"]

# Rank states by per-capita mortality
state_stats["rank"] = state_stats["per_capita_mortality"].rank(ascending=False)
print(state_stats[["per_capita_mortality", "rank"]].sort_values("rank"))
```

Here, I used COVID-19 data to compute and rank U.S. states based on per-capita mortality during the period from January 1, 2020, to February 28, 2021. First, I converted the date column into a datetime format to facilitate time-based filtering. The dataset was then restricted to the specified time range, ensuring that only relevant records were analyzed. The code grouped the data by state, calculating the total number of deaths and the average population for each state. Using these values, per-capita mortality was computed as the ratio of total deaths to the average population, providing a normalized measure of mortality risk across states. The states were then ranked in descending order based on per-capita mortality, with lower ranks indicating higher mortality rates. The results indicate that New York experienced the highest per-capita mortality during this period, followed by Michigan, Louisiana, and Illinois.

During the same time period, compute the case fatality rate (CFR) per month, using one of the approaches defined in this scientific brief from the World Health Organization: https://www.who.int/news-room/commentaries/detail/estimating-mortality-from-covid-19Links to an external site.. This computation should yield an array of dimensions 50 (states) X 14 (months). State the assumptions you are making in computing this metric.

```python
# Extract year and month from the date column
time_period_df["date"] = dd.to_datetime(time_period_df["date"])
time_period_df["month"] = time_period_df["date"].dt.to_period("M")

# Group by state and month, and compute total deaths and cases
monthly_cfr = time_period_df.groupby(["state", "month"]).agg(
    total_deaths=("deaths", "sum"),
    total_cases=("cases", "sum")
).compute()

# Compute CFR
monthly_cfr["cfr"] = (monthly_cfr["total_deaths"] / monthly_cfr["total_cases"]) * 100

cfr_matrix = monthly_cfr.pivot_table(index="state", columns="month", values="cfr")
cfr_matrix = cfr_matrix.fillna(0)
print("Case Fatality Rate Matrix:")
cfr_matrix
```

In computing the Case Fatality Rate (CFR) per month for each U.S. state, I have assumed that the likelihood of detecting COVID-19 cases and deaths remains consistent over the outbreak period. Early in the pandemic, surveillance systems prioritized symptomatic individuals seeking care, leading to a potential overestimation of CFR as asymptomatic and mild cases were underreported. As testing capacity and case definitions improved, case detection likely became more comprehensive, reducing this bias over time. To mitigate early-stage overestimation, one potential approach is left censoring removing cases from the analysis that occurred before the establishment of robust surveillance systems. The generated CFR matrix, structured with states as rows and months as columns, allows for trend analysis of case fatality rates across different regions and timeframes, helping to identify shifts in mortality risk and healthcare response effectiveness.

4. Using this matrix as input, compute the ranking of states, on how the CFR rate changed over time. This computation will involve an aggregation of of month-to-month changes in CFR. Note that some of these month-to-month changes can be positive (CFR increases from a previous month to the current month) or negative (CFR decrease from a previous month to the current month), or zero (no change in CFR). You need to aggregate these individual changes, across all time periods.

```python
# Compute month-to-month changes in CFR
cfr_matrix_diff = cfr_matrix.diff(axis=1)  
cfr_change_aggregated = cfr_matrix_diff.sum(axis=1)
cfr_change_aggregated = cfr_change_aggregated.sort_values(ascending=False)
print("cfr_change_aggregated:")
cfr_change_aggregated
```

Interpretation: The month-to-month changes in Case Fatality Rate (CFR) reveal significant variations across states, with some experiencing substantial increases while others remained relatively stable. Connecticut, Michigan, and New Jersey saw the highest CFR changes, exceeding 8.0, followed closely by Massachusetts, Pennsylvania, and New York. The Northern Mariana Islands and U.S. Virgin Islands also exhibited notable fluctuations, suggesting considerable month-to-month variability. States such as Indiana, Ohio, Louisiana, and Colorado had moderate CFR changes, ranging between 5.0 and 6.0, while others like Florida, North Carolina, and Kansas saw smaller increases, between 1.5 and 2.5. Interestingly, Hawaii reported no change in CFR, indicating complete stability during the analyzed period.
